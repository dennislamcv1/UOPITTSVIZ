## Step 2: Required Plots


### Line Plot:
Plot average temperature against the months.
Customize the plot with a title, axis labels, and grid lines.


```python
# Line Plot - Average Temperature Over Months
plt.figure(figsize=(8, 5))
plt.plot(months, temperatures, marker='o', linestyle='-', color='b')
plt.title('Average Monthly Temperature')
plt.xlabel('Months')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

Create a bar chart showing ice cream and coffee sales side by side for each month.
Use a legend to differentiate between the two.
