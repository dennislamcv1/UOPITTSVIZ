```python
#preamble

import seaborn as sns
import matplotlib.pyplot as plt
```


## Task 1: Load and explore the dataset
Use built-in dataset of flights:

```python
# loading dataset of flights
df = sns.load_dataset('flights')
df.head()
df.describe()
```

## Task 2: Data visualization and analysis
* Create a scatter plot to depict number of passengers over the years, with months represented as a hue
* Write up your observation about changes in passenger rates over time and in different months


```python
sns.relplot(data=df,y='passengers',x='year',hue="month")
```

## Additional exploration of the data
Ask the user to enter two specifc years.  Create a subset of the data including the data for those specifc years. Plot the scatterlots for these two years to compare number of passengers over months in each year

```python
# Get user input
year1 = int(input("Enter a year between 1949 and 1960: "))
year2 = int(input("Enter another year for comparison: "))
```

```python
# Validate input
if year1 < 1949 or year1 > 1960 or year2 < 1949 or year2 > 1960:
    print("Error: Please enter years between 1949 and 1960.")
elif year1 == year2:
    print("Error: The two years must be different.")
else:
    # Filter the dataset for the specific years
    specific_years = [year1, year2]
    flights_specific_years = df[df['year'].isin(specific_years)]
    print(f"Filtered data for the years {year1} and {year2}:")
    print(flights_specific_years)
```

```python
sns.catplot(data=flights_specific_years,x='month',y='passengers',col='year')
#sns.relplot(data=df,y='passengers',x='year',hue="month")
```
