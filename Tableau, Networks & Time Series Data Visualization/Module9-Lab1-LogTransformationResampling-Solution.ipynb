### **Lab Instructions**

#### **1. Import Required Libraries**
Start by importing the necessary Python libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### **2. Load the Dataset**
Use a sample dataset or generate synthetic time series data.

```python
#load the water usage dataset
df = pd.read_csv('Water_Usage_Data.csv', parse_dates=['Date'], index_col='Date')
# Display the first ten rows of the total_gallons data
print(df['total_gallons'].head(10))
```

#### **3. Apply Log Transformation**
Transform the time series using the natural logarithm to stabilize variance.

```python
# Apply log transformation
df['Log_Value'] = np.log(df['total_gallons'])

#display the first few rows of the transformmed data
print(df['Log_Value'].head(10))
```

#### **4 Resampling the Time Series**
Aggregate the data to a monthly frequency and compute the mean for each month.


```python
# Resample to monthly frequency
df_monthly = df['total_gallons'].resample('M').mean()

#display the first few rows of the resampled data
print(df_monthly.head(10))
```

#### **5. Combine Log Transformation and Resampling**
Apply the log transformation to the original data and then resample it.


```python
# Resample the log-transformed data
df_log_monthly = df['Log_Value'].resample('M').mean()

#display the first few rows of the resampled data
print(df_log_monthly.head(10))
```
