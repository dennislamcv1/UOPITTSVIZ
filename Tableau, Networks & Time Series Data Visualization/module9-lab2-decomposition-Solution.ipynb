#### **2. Load the Dataset**
Use a sample dataset or generate synthetic time series data.

```python
#load the water usage dataset
df = pd.read_csv('Water_Usage_Data.csv', parse_dates=['Date'], index_col='Date')
# Display the first few rows of the dataset
print(df.head())
```

#### **3. Perform seasonal decomposition*
Using decompose function of the statsmodels library, decompose the water usage of multi family households into its component

```python
#import decomposing function from statsmodels library

from statsmodels.tsa.seasonal import seasonal_decompose 

#decompose the water usage of multi family households into its components
decomposition = seasonal_decompose(df['multi_family_gallons'], model='additive', period=12) 

# Plot the components 
decomposition.plot()
```
